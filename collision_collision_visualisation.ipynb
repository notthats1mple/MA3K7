```python
import random
import matplotlib.pyplot as plt
num_A = 10
num_B = 500
num_C = 3000
history_A = [num_A]
history_B = [num_B]
history_C = [num_C]

def collide_and_transform(num_A, num_B, num_C):
    particles = ['A'] * num_A + ['B'] * num_B + ['C'] * num_C
    random.shuffle(particles)
    particle1, particle2 = random.sample(particles, 2)
    if {particle1, particle2} == {'A', 'B'}:
        num_A -= 1
        num_B -= 1
        num_C += 2
    elif {particle1, particle2} == {'A', 'C'}:
        num_A -= 1
        num_C -= 1
        num_B += 2
    elif {particle1, particle2} == {'B', 'C'}:
        num_B -= 1
        num_C -= 1
        num_A += 2  
    return num_A, num_B, num_C

num_collisions = 50000  # Number of collisions to simulate
for _ in range(num_collisions):
    num_A, num_B, num_C = collide_and_transform(num_A, num_B, num_C)
    history_A.append(num_A)
    history_B.append(num_B)
    history_C.append(num_C)
    
    
history_A_first_1000 = history_A[0:1000]
history_B_first_1000 = history_B[0:1000]
history_C_first_1000 = history_C[0:1000]

# Generate the plot for the first 1000 collisions
plt.plot(range(0, 1000), history_A_first_1000, label='A Particles')
plt.plot(range(0, 1000), history_B_first_1000, label='B Particles')
plt.plot(range(0, 1000), history_C_first_1000, label='C Particles')
plt.legend()
plt.title('Particle Populations Over Time (First 1000 Collisions)')
plt.xlabel('Collision Number')
plt.ylabel('Population Size')
plt.show()
```

```python
history_A_last_1000 = history_A[49000:50000]
history_B_last_1000 = history_B[49000:50000]
history_C_last_1000= history_C[49000:50000]

# Generate the plot for the last 1000 collisions
plt.plot(range(49000, 50000), history_A_last_1000, label='A Particles')
plt.plot(range(49000, 50000), history_B_last_1000, label='B Particles')
plt.plot(range(49000, 50000), history_C_last_1000, label='C Particles')
plt.legend()
plt.title('Particle Populations Over Time (Last 1000 Collisions)')
plt.xlabel('Collision Number')
plt.ylabel('Population Size')
plt.ylim(500,2000)
plt.show()
```
